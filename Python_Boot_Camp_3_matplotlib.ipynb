```
import numpy as np
import matplotlib.pyplot as plt
```

# BASICS


```
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```
# COLORS!
x = np.linspace(0, 10, 50)
plt.plot(x, np.sin(x-0), color='blue')
plt.plot(x, np.sin(x-1), color='g')  # rgbcmyk
plt.plot(x, np.sin(x-2), color='0.75')  # grayscale
plt.plot(x, np.sin(x-3), color='#FFDD44')  # grayscale
plt.plot(x, np.sin(x-4), color=(1,0.2, 0.3))  # rgb values
plt.show()
```

```
#LINES!
x = np.linspace(0, 10, 50)
plt.plot(x, x+1, ls='solid')
plt.plot(x, x+2, ls='dotted')
plt.plot(x, x+3, ls='dashed')
plt.plot(x, x+4, ls='-')
plt.plot(x, x+5, ls='--')
plt.plot(x, x+6, ls='-.')
plt.plot(x, x+7, ls=':')
plt.show()
```

```
# STEVE!
x = np.linspace(1,10, 10)
y = np.log10(x)
plt.plot(x,y, color='green', ls='dashed', linewidth=2, marker='o', markersize=12)
plt.plot(x,y+0.5, 'r*:')
plt.show()
```

```
# RECAP
x = np.linspace(0, 10, 25)
y1 = np.cos(x)
offset = 0.5
y2 = y1 + offset
y3 = y2 + offset
y4 = y3 + offset

plt.plot(x,y1,'b-')   # blue line
plt.plot(x,y2,'r:')   # red dotted line
plt.plot(x,y3,'gs')   # green squares (no line)
plt.plot(x,y4,'mo--') # magenta points and a dashed line
plt.show()
```

#PROPER PLOTS


```
plt.figure(figsize=(8,5))
x = np.linspace(0, 10)

plt.plot(x, np.sin(x), 'g-', label='sin')
plt.plot(x, np.cos(x), 'b:', label='cos')

plt.xlim(1,8)
plt.ylim(-1, 2)

plt.title('A Proper Plot')
plt.xlabel('Time [s]')
plt.ylabel('Distance [m]')

plt.grid()
plt.legend()
plt.show()
```

#HISTOGRAMS


```
data = np.random.randn(1000)
plt.hist(data)
plt.show()
```

```
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

plt.hist(x1, histtype='stepfilled', alpha=0.3, density=True, bins=10)
plt.hist(x2, histtype='stepfilled', alpha=0.3, density=True, bins=10)
plt.hist(x3, histtype='stepfilled', alpha=0.3, density=True, bins=10)
plt.show()
```

# SUBPLOTS


```
plt.figure(figsize=(12,4))

x = np.linspace(0,10,10)

plt.subplot(1,3,1)
plt.plot(x, np.sin(x), 'r-')
plt.title('SIN')
plt.ylabel('units')

plt.subplot(1,3,2)
plt.plot(x, np.cos(x), 'g--')
plt.title('COS')
plt.ylabel('units')

plt.subplot(1,3,3)
plt.plot(x, np.exp(x), 'b^')
plt.title('EXP')
plt.ylabel('units')


plt.tight_layout()
plt.show()
```
